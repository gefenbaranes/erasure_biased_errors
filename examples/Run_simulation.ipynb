```python
real_losses_by_instruction_ix = {12: 2, 30: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {14: 2, 30: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {16: 2, 30: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {12: 2, 34: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {14: 2, 34: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {16: 2, 34: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {12: 2, 46: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {14: 2, 46: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {16: 2, 46: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {12: 2, 50: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {14: 2, 50: 4} # {instruction_ix: qubit, ..}
real_losses_by_instruction_ix = {16: 2, 50: 4} # {instruction_ix: qubit, ..}

real_losses_by_instruction_ix_list = [{12: 2, 30: 4}, {14: 2, 30: 4}, {16: 2, 30: 4} ,
                                    {12: 2, 34: 4}, {14: 2, 34: 4}, {16: 2, 34: 4} ,
                                    {12: 2, 46: 4}, {14: 2, 46: 4}, {16: 2, 46: 4} ,
                                    {12: 2, 50: 4}, {14: 2, 50: 4}, {16: 2, 50: 4} ,
                                    ]



from Simulator import *
from HeraldedCircuit_SWAP_LD import HeraldedCircuit_SWAP_LD
from HeraldedCircuit_MLE_LD import HeraldedCircuit_MLE_LD
from HeraldedCircuit_FREE_LD import HeraldedCircuit_FREE_LD
from distutils.util import strtobool


### Step 1: choose input parameter:
Meta_params = {'architecture': 'CBQC', 'code': 'Rotated_Surface', 'bias_preserving_gates': 'True', 'logical_basis': 'Z', 'noise': 'biased_erasure_noise', 'is_erasure_biased': 'False',
            'LD_freq': 1, 'SSR': 'True', 'LD_method': 'SWAP', 'ordering': 'good', 'decoder': 'MLE', 'num_logicals': 1, 'circuit_type': 'memory', 
            'cycles': 4, 'Steane_type': None, 'printing': 'False'}
bloch_point_params = {'bias_ratio': 0.5, 'erasure_ratio': 1.0}
phys_error = 1e-2
distance = 2
real_losses_by_instruction_ix = {14: 2, 30: 4} # {instruction_ix: qubit, ..}
for real_losses_by_instruction_ix in real_losses_by_instruction_ix_list:


    ### Step 2: generate the clean circuit:
    noise_dict = {'biased_erasure_noise': biased_erasure_noise, 'no_noise': no_noise, 'biased_erasure_noise_correlated': biased_erasure_noise_correlated, 'atom_array': atom_array}
    noise = noise_dict[Meta_params['noise']]
    atom_array_sim = True if Meta_params['noise'] == 'atom_array' else False
    loss_detection_method_dict = {'FREE': HeraldedCircuit_FREE_LD, 'SWAP': HeraldedCircuit_SWAP_LD}
    loss_detection_method = loss_detection_method_dict[Meta_params['LD_method']]

    if Meta_params['code'] not in ['Rotated_XZZX', 'Rotated_Surface']: 
        Meta_params['ordering'] = None # not relevant
        
    cycles_str = Meta_params['cycles']; cycles = None if (cycles_str == 'd' or cycles_str is None) else int(cycles_str)

    simulator = Simulator(architecture=Meta_params['architecture'], code=Meta_params['code'],
                    bias_preserving_gates=strtobool(Meta_params['bias_preserving_gates']), logical_basis=Meta_params['logical_basis'],
                    noise=noise, atom_array_sim=atom_array_sim ,is_erasure_biased=strtobool(Meta_params['is_erasure_biased']), 
                    bias_ratio=float(bloch_point_params['bias_ratio']), erasure_ratio=float(bloch_point_params['erasure_ratio']),
                    loss_detection_freq=int(Meta_params['LD_freq']), SSR=strtobool(Meta_params['SSR']),
                    loss_detection_method=loss_detection_method, cycles = cycles, loss_detection_method_str = Meta_params['LD_method'],
                    ordering_type=Meta_params['ordering'], decoder = Meta_params['decoder'], num_logicals=int(Meta_params['num_logicals']), 
                    circuit_type=Meta_params['circuit_type'], Steane_type=Meta_params['Steane_type'], printing=strtobool(Meta_params['printing']))

    LogicalCircuit = simulator.generate_circuit(d=distance, cycles=cycles, phys_err=phys_error)

    ancilla_qubits = [qubit for i in range(int(Meta_params['num_logicals'])) for qubit in LogicalCircuit.logical_qubits[i].measure_qubits]
    data_qubits = [qubit for i in range(int(Meta_params['num_logicals'])) for qubit in LogicalCircuit.logical_qubits[i].data_qubits]
            
            
    HeraldedCircuit_SWAP_LD_class = HeraldedCircuit_SWAP_LD(circuit=LogicalCircuit, biased_erasure=strtobool(Meta_params['is_erasure_biased']), 
                                                            bias_preserving_gates=strtobool(Meta_params['bias_preserving_gates']),
                                                                basis = Meta_params['logical_basis'], distance=distance, erasure_ratio = float(bloch_point_params['erasure_ratio']), 
                                                                phys_error = phys_error, data_qubits=data_qubits, ancilla_qubits=ancilla_qubits,
                                                                SSR=strtobool(Meta_params['SSR']), cycles=cycles, printing=strtobool(Meta_params['printing']), 
                                                                loss_detection_freq = int(Meta_params['LD_freq']))

    HeraldedCircuit_MLE_LD_class = HeraldedCircuit_MLE_LD(circuit=LogicalCircuit, biased_erasure=strtobool(Meta_params['is_erasure_biased']), 
                                                        bias_preserving_gates=strtobool(Meta_params['bias_preserving_gates']),
                                                                basis = Meta_params['logical_basis'], distance=distance, erasure_ratio = float(bloch_point_params['erasure_ratio']), 
                                                                phys_error = phys_error, data_qubits=data_qubits, ancilla_qubits=ancilla_qubits,
                                                                SSR=strtobool(Meta_params['SSR']), cycles=cycles, printing=strtobool(Meta_params['printing']), 
                                                                loss_detection_freq = int(Meta_params['LD_freq']))


    SWAP_circuit = HeraldedCircuit_SWAP_LD_class.transfer_circuit_into_SWAP_circuit(LogicalCircuit)


    # print(f"The clean SWAP circuit is: \n{SWAP_circuit}")
        
        
    HeraldedCircuit_MLE_LD_class.circuit = SWAP_circuit
    HeraldedCircuit_MLE_LD_class.set_up_Pauli_DEM()
    HeraldedCircuit_MLE_LD_class.get_qubits_lifecycles() 
    HeraldedCircuit_MLE_LD_class.rounds_by_ix = HeraldedCircuit_MLE_LD_class.split_stim_circuit_into_rounds()


    ### Step 5: generate the experimental circuit:
    lost_qubits = set(real_losses_by_instruction_ix.values())
    first_loss_instruction_index = min(real_losses_by_instruction_ix)
    loss_qubits_remove_gates_range = {}
    for instruction_ix in real_losses_by_instruction_ix:
        lost_q = real_losses_by_instruction_ix[instruction_ix]
        round_ix = next((round_ix for round_ix, instructions in sorted(HeraldedCircuit_MLE_LD_class.rounds_by_ix.items()) if sum(len(HeraldedCircuit_MLE_LD_class.rounds_by_ix[r]) for r in range(-1, round_ix+1)) > instruction_ix), None)

        [reset_round_ix, detection_round_ix] = next(([cycle[0],cycle[1]] for cycle in HeraldedCircuit_MLE_LD_class.qubit_lifecycles_and_losses[lost_q] if cycle[0] <= round_ix <= cycle[1]), None)
        # detection_round_offset_start = sum(len(self.rounds_by_ix[round_ix]) for round_ix in self.rounds_by_ix if round_ix < reset_round_ix)
        detection_round_offset_end = sum(len(HeraldedCircuit_MLE_LD_class.rounds_by_ix[round_ix]) for round_ix in HeraldedCircuit_MLE_LD_class.rounds_by_ix if round_ix < detection_round_ix + 1)
        
        if lost_q in loss_qubits_remove_gates_range: # this qubit was already recorder for loss, and was lost again!
            loss_qubits_remove_gates_range[lost_q].append((instruction_ix, detection_round_offset_end))
        else:
            loss_qubits_remove_gates_range[lost_q] = [(instruction_ix, detection_round_offset_end)]

    circuit_before_ix = HeraldedCircuit_MLE_LD_class.circuit[:first_loss_instruction_index]
    circuit_after_ix = HeraldedCircuit_MLE_LD_class.circuit[first_loss_instruction_index:]


    print(f"\nLets generate the real experimental circuit with the loss event of qubits {lost_qubits} in locations {real_losses_by_instruction_ix.keys()}")
    heralded_circuit_after_ix = HeraldedCircuit_MLE_LD_class.generate_circuit_without_lost_qubit(lost_qubits = lost_qubits, circuit = circuit_after_ix, circuit_offset = first_loss_instruction_index,loss_qubits_remove_gates_range=loss_qubits_remove_gates_range) # after removing the following gates with the lost qubits.

    experimental_circuit = circuit_before_ix + heralded_circuit_after_ix

    # print(f"\nThe experimental circuit is: {experimental_circuit}")
                
    # replace final observables with detectors:
    # WARNING: we assume that the OBSERVABLES are at the end of the circuit.
    observable_instructions_ix = [i for i, instruction in enumerate(experimental_circuit) if instruction.name == 'OBSERVABLE_INCLUDE']
    extra_num_detectors = len(observable_instructions_ix)
    final_heralded_circuit = experimental_circuit[:-extra_num_detectors].copy()
    for obs_ix in observable_instructions_ix:
        instruction = experimental_circuit[obs_ix]
        targets = instruction.targets_copy()
        final_heralded_circuit.append('DETECTOR', targets)
    # get the dem of the experimental circuit:
    dem_heralded_circuit = final_heralded_circuit.detector_error_model(decompose_errors=False, approximate_disjoint_errors=True, ignore_decomposition_failures=True, allow_gauge_detectors=True)
    print(f"The DEM for the experimental circuit is: \n{dem_heralded_circuit}")
```

```python
import numpy as np
import qec
import pymatching, stim
import matplotlib.pyplot as plt
import statsmodels.stats.proportion as stat


def memory(d, p=0.005):
    logical_qubits = [qec.surface_code.RotatedSurfaceCode(d, d)]
    lc = qec.LogicalCircuit(logical_qubits, initialize_circuit=False,
                            loss_noise_scale_factor=0.0, spam_noise_scale_factor=1,
                            gate_noise_scale_factor=1, idle_noise_scale_factor=0,
                            measurement_error_rate=0,
                            reset_error_rate=0,
                            entangling_error_rate=(p / 3, p / 3, p / 3),
                            single_qubit_error_rate=(0, 0, 0))

    # First prepare all logical qubits in |0>
    lc.append(qec.surface_code.prepare_zero, list(range(0, len(logical_qubits))))

    for _ in range(d-1):
        lc.append(qec.surface_code.measure_stabilizers, [0], move_duration=0)

    lc.append(qec.surface_code.measure_z, list(range(len(logical_qubits))), observable_include=False)
    lc.append('MOVE_TO_NO_NOISE', lc.qubit_indices, 0)

    lc.append('MPP', lc.logical_qubits[0].logical_z_operator)
    lc.append('OBSERVABLE_INCLUDE', [stim.target_rec(-1)], lc.num_observables)

    return lc


def memory_for_preselection(d, p=0.005):
    logical_qubits = [qec.surface_code.RotatedSurfaceCode(d, d)]
    lc = qec.LogicalCircuit(logical_qubits, initialize_circuit=False,
                            loss_noise_scale_factor=0.0, spam_noise_scale_factor=1,
                            gate_noise_scale_factor=1, idle_noise_scale_factor=0,
                            measurement_error_rate=0,
                            reset_error_rate=0,
                            entangling_error_rate=(p, p, p),
                            single_qubit_error_rate=(0, 0, 0))

    # First prepare all logical qubits in |0>
    lc.append(qec.surface_code.prepare_zero, list(range(0, len(logical_qubits))))

    for _ in range(d//2):
        lc.append(qec.surface_code.measure_stabilizers, [0], move_duration=0, compare_with_previous=True)

    lc.spam_noise_scale_factor = 0
    lc.gate_noise_scale_factor = 0

    lc.append('MPP', lc.logical_qubits[0].logical_z_operator)
    lc.append('OBSERVABLE_INCLUDE', [stim.target_rec(-1)], lc.num_observables)

    lc.spam_noise_scale_factor = 0
    lc.gate_noise_scale_factor = 1

    for _ in range(d//2):
        lc.append(qec.surface_code.measure_stabilizers, [0], move_duration=0, compare_with_previous=True)
    return lc


fidelity = 0
num_shots = 1000
ds = [3, 5]
acceptance_fractions = np.linspace(0, 1, 21)[:-1]

colors = ['#4287f5', '#9c6df2', '#79f2d6', '#e86da2']
edgecolors = ['#1d66db', '#7639e6', '#3ec9a9', '#ab3769']



for (_, d) in enumerate(ds):
    print('Memory with d = {}'.format(d))
    lc = memory(d)
    lc_for_preselection = memory_for_preselection(d)

    dem = lc.without_loss().detector_error_model(decompose_errors=False, approximate_disjoint_errors=True)
    dem_for_preselection = lc_for_preselection.without_loss().detector_error_model(decompose_errors=False,
                                                                                   approximate_disjoint_errors=True)

    # Sample the circuit.
    shots = lc.compile_sampler().sample(num_shots)
    detectors, observables = lc.without_loss().compile_m2d_converter().convert(measurements=shots,
                                                                               separate_observables=True)

    corrections_for_preselection, probabilities = qec.correlated_decoders.mle_sliding_scale.sliding_scale(dem_for_preselection,
                                                                                                          detectors[:, :dem_for_preselection.num_detectors])
    corrections = qec.correlated_decoders.mle.decode_gurobi_with_dem(dem, detectors)
    np.save('corrections_memory_{}.npy'.format(d), corrections)
    np.save('observables_memory_{}.npy'.format(d), observables)
    np.save('probabilities_memory_{}.npy'.format(d), probabilities)

    corrected_observables = 1 - np.logical_xor(observables.flatten(), corrections.flatten())
    p0 = np.e ** probabilities[:, 0] / (np.e ** probabilities[:, 0] + np.e ** probabilities[:, 1])
    p1 = np.e ** probabilities[:, 1] / (np.e ** probabilities[:, 0] + np.e ** probabilities[:, 1])
    complimentary_gaps = p0 * np.log(p0) + p1 * np.log(p1)
    index_order = np.argsort(complimentary_gaps)

    complimentary_gaps = complimentary_gaps[index_order]
    probabilities = probabilities[index_order]
    corrected_observables = corrected_observables[index_order]




    # Plot results
    fidelities = [1 - np.mean(corrected_observables[int(len(complimentary_gaps) * acceptance_fractions[i]):]) for i in
                  range(len(acceptance_fractions))]
    lbs = []
    ubs = []
    for i in range(len(acceptance_fractions)):
        lb, ub = stat.proportion_confint(
            np.sum(1 - corrected_observables[int(len(complimentary_gaps) * acceptance_fractions[i]):]),
            len(corrected_observables[int(len(complimentary_gaps) * acceptance_fractions[i]):]), method='beta',
            alpha=.32)
        lbs.append((fidelities[i]) - lb)
        ubs.append(ub - fidelities[i])

    lbs = np.array(lbs)
    ubs = np.array(ubs)
    fidelities = np.array(fidelities)
    plt.errorbar((1 - acceptance_fractions)[fidelities > 0], fidelities[fidelities > 0],
                 yerr=[lbs[fidelities > 0], ubs[fidelities > 0]],
                 marker=None, color=colors[_], markeredgecolor=edgecolors[_])
    plt.plot((1 - acceptance_fractions)[fidelities > 0], fidelities[fidelities > 0],
             marker='o', color=colors[_], markeredgecolor=edgecolors[_], label='$d={}$'.format(d), zorder=100)


plt.semilogy()
plt.legend(frameon=False)
plt.xlabel('Acceptance fraction')
plt.ylabel('Logical error rate')
plt.show()


```
